first we have to check if their are any missing values or not, if found then we have to sort it.

```python
dataset.isnull().sum()
```

so we have found that there are 72 null values in product base margin
and now we have to fill it

```python
#to replace missing value we use fill na with mean() function and repacing permanently we use inplace

dataset['Product Base Margin'].fillna(dataset['Product Base Margin'].mean(),inplace = True) 
```

```python
#now lets again check if its filled or not

dataset.isnull().sum()
```

# Order Priority

```python


#we have to check the priority details here in this column
#so we use value_counts()

dataset['Order Priority'].value_counts()
```

```python
#now we we will make graph using seaborn library
# as its a good and easy way to make client understand using graphs
plt.figure(figsize=(5,4)) #making the size of graph small

sns.countplot(x="Order Priority",data = dataset)
plt.title("Count of Order Priority")
plt.savefig("Count of Order Priority") #you can save graph to use it in excel,powerpoint,word for your presentation
plt.show()
```

# Ship Mode

```python
#first of all we have to check from in which sipping mode our products are sales
```

```python
dataset["Ship Mode"].value_counts()
```

```python
x = dataset['Ship Mode'].value_counts().index
y = dataset['Ship Mode'].value_counts().values
```

```python
plt.figure(figsize=(5,4))
#fixing the size of pie chart

plt.pie(y,labels=x,startangle=60,autopct="%0.2f%%")
#creating pie chart using pie function and giving its parameter to adjust, autopct used to give percentage in the graph


plt.legend(loc = 2)# as legends shows a basic properties to view graph in an easy way
plt.show()
```

This two are univariate analysis as we see in above codes

# Bivariate Analysis

```python
plt.figure(figsize=(4,3))
sns.countplot(x="Ship Mode",data=dataset,hue="Product Category")
plt.show()
```

so here we can see that more numbers of office supplies are shipped through regular air mode.
